```python
import numpy as np
import matplotlib.pyplot as plt

class BackdoorDetector:
    def __init__(self, models):
        self.models, self.alerts = models, []
    
    def scan_portfolio(self):
        print("üîç Scanning 50+ model portfolio...")
        
        for i in range(len(self.models)):
            # Simulate activation analysis
            activations = np.random.normal(1, 0.2, 10)  # 10 classes
            activations[7] *= 2.5  # Class 7 anomaly
            
            if activations[7] > 2.0:
                self.alerts.append(f"Model {i}: Class 7 anomaly")
                print(f"üö® Model {i}: Class 7 activation = {activations[7]:.2f}")
        
        print(f"üìä Scan complete: {len(self.alerts)} backdoors detected")
        return self.alerts

def plot_detection():
    classes = range(10)
    normal_acts = np.random.normal(1, 0.1, 10)
    backdoor_acts = normal_acts.copy()
    backdoor_acts[7] = 2.7  # Anomalous class 7
    
    plt.bar(classes, normal_acts, alpha=0.7, label='Normal')
    plt.bar(classes, backdoor_acts, alpha=0.7, label='Backdoor')
    plt.axhline(y=2.0, color='r', linestyle='--', label='Alert threshold')
    plt.xlabel('Classes'); plt.ylabel('Activation'); plt.legend(); plt.show()

if __name__ == "__main__":
    detector = BackdoorDetector([None]*50)  # 50 dummy models
    alerts = detector.scan_portfolio()
    plot_detection()
```
