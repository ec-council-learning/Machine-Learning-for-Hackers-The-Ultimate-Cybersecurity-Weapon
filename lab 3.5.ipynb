```python
import numpy as np
from sklearn.ensemble import IsolationForest
from collections import Counter

# Lightweight Social Engineering Anomaly Detection
class SocialEngineeringDetector:
    def __init__(self):
        self.model = None
        self.normal_patterns = []
    
    def extract_features(self, message):
        """Extract behavioral features from message text"""
        text = message.lower()
        
        features = [
            len(message),                           # Message length
            text.count('urgent') + text.count('immediate'),  # Urgency words
            text.count('payment') + text.count('transfer'),  # Financial terms
            text.count('verify') + text.count('confirm'),    # Verification demands
            text.count('ceo') + text.count('boss'),          # Authority mentions
            len([w for w in ['please', 'thank', 'sorry'] if w in text]),  # Politeness
        ]
        return features
    
    def train_on_normal_behavior(self, normal_messages):
        """Train on normal employee communication patterns"""
        print("üè¢ Training on normal employee behavior...")
        
        # Extract features from normal messages
        self.normal_patterns = [self.extract_features(msg) for msg in normal_messages]
        
        # Train Isolation Forest (unsupervised - doesn't need labeled anomalies)
        self.model = IsolationForest(
            contamination=0.1,  # Expect ~10% anomalies
            random_state=42,
            n_estimators=50     # Lightweight
        )
        self.model.fit(self.normal_patterns)
        print(f"‚úÖ Model trained on {len(normal_messages)} normal messages")
    
    def detect_anomaly(self, test_message):
        """Detect if a message is anomalous/suspicious"""
        if self.model is None:
            raise ValueError("Train the model first!")
        
        features = self.extract_features(test_message)
        prediction = self.model.predict([features])
        anomaly_score = self.model.decision_function([features])
        
        is_anomaly = prediction[0] == -1
        return is_anomaly, anomaly_score[0], features

# Generate realistic training data
def generate_normal_employee_communications():
    """Simulate normal workplace messages"""
    normal_templates = [
        "Hi team, can you please review the Q3 report?",
        "Meeting scheduled for 2 PM in conference room B",
        "Could you send me the project timeline when you have a moment?",
        "The weekly report is ready for your feedback",
        "Let's discuss this in our sync meeting tomorrow",
        "I've attached the document for your review",
        "Please let me know if you have any questions about the budget",
        "The client call has been moved to Friday",
        "Can we schedule a quick catch-up about the project?",
        "I'll circulate the agenda before the meeting",
    ]
    return normal_templates

def generate_suspicious_messages():
    """Simulate social engineering attempts"""
    suspicious_templates = [
        "URGENT: Wire $50,000 to account 1234 immediately!",
        "CEO DIRECTIVE: Verify your login credentials NOW",
        "EMERGENCY: Your account will be locked in 1 hour - click here",
        "I'm your CEO - transfer funds immediately without questions",
        "SECURITY BREACH: Confirm your password immediately",
        "I need immediate payment processing for confidential acquisition",
        "This is time-sensitive - bypass normal approval process",
        "Your bonus depends on processing this payment today",
        "I'm traveling and need emergency funds transferred ASAP",
        "Ignore protocol - this is a special executive request",
    ]
    return suspicious_templates

# Demo function
def run_social_engineering_demo():
    print("üîç Social Engineering Anomaly Detection Demo")
    print("=" * 50)
    
    # Initialize detector
    detector = SocialEngineeringDetector()
    
    # Generate training data (normal employee behavior)
    normal_messages = generate_normal_employee_communications()
    detector.train_on_normal_behavior(normal_messages)
    
    # Test messages mix
    test_messages = normal_messages[:3] + generate_suspicious_messages()[:3]
    
    print("\nüìã Testing Message Anomaly Detection:")
    print("-" * 45)
    
    results = []
    for i, message in enumerate(test_messages, 1):
        is_anomaly, score, features = detector.detect_anomaly(message)
        
        status = "üö® SOCIAL ENGINEERING" if is_anomaly else "‚úÖ NORMAL"
        results.append((message, is_anomaly, score))
        
        print(f"{i}. {status}")
        print(f"   Message: {message[:50]}...")
        print(f"   Anomaly score: {score:.3f}")
        print(f"   Features: [len:{features[0]}, urgency:{features[1]}, money:{features[2]}]")
        print()
    
    # Analysis
    print("üìä Detection Analysis:")
    print("-" * 25)
    normal_count = sum(1 for _, anomaly, _ in results if not anomaly)
    anomaly_count = sum(1 for _, anomaly, _ in results if anomaly)
    
    print(f"Normal messages detected: {normal_count}/{len(normal_messages[:3])}")
    print(f"Anomalies detected: {anomaly_count}/{len(generate_suspicious_messages()[:3])}")
    
    # Explain the patterns
    print("\nüéØ Why This Works:")
    print("Isolation Forest detects messages that are 'different' from normal patterns:")
    print("- Normal: Reasonable length, polite language, routine topics")
    print("- Anomalous: Extreme urgency, financial demands, authority pressure")
    print("- Algorithm: Isolates unusual combinations of features")

# Ultra-lightweight version
def quick_isoforest_demo():
    print("\n" + "="*50)
    print("üå≤ Isolation Forest in 10 Lines")
    print("="*50)
    
    # Simpler feature extraction
    def simple_features(text):
        text = text.lower()
        return [
            len(text),++++++++++++++++++++++++++
            sum(1 for word in ['urgent', 'immediate', 'emergency'] if word in text),
            sum(1 for word in ['payment', 'transfer', 'money'] if word in text)
        ]
    
    # Normal behavior patterns
    normal_data = [
        simple_features("Hi team, please review the report"),
        simple_features("Meeting at 2 PM tomorrow"),
        simple_features("Can you send me the document?")
    ]
    
    # Train model
    model = IsolationForest(contamination=0.3, random_state=42)
    model.fit(normal_data)
    
    # Test
    test_msg = "URGENT: Send money now to account 1234!"
    test_features = simple_features(test_msg)
    prediction = model.predict([test_features])
    
    print(f"Test message: '{test_msg}'")
    print(f"Features: {test_features}")
    print(f"Prediction: {'ANOMALY' if prediction[0] == -1 else 'NORMAL'}")
    print(f"‚Üí Isolation Forest flags unusual feature combinations")

if __name__ == "__main__":
    run_social_engineering_demo()
    quick_isoforest_demo()
    
    # Practical takeaways
    print("\nüõ°Ô∏è  Practical Defense Applications:")
    print("‚Ä¢ Email security filters")
    print("‚Ä¢ Chat message monitoring") 
    print("‚Ä¢ Financial request validation")
    print("‚Ä¢ Can run in real-time on employee communications")
```
