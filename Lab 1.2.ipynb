# Video 1.2 – Learn to Interact with ML Models Using Python

In this notebook, we will:

1. **Define** a simple neural network using *PyTorch*
2. **Pass** input data and get predictions
3. **Inspect** model structure, weights, and biases

```python
# Import required PyTorch libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
```

## 1 – Define a Simple Neural Network

```python
# Define a simple feed‑forward neural network
# - Input layer: 4 features
# - Hidden layer: 16 neurons, ReLU activation
# - Output layer: 3 classes
model = nn.Sequential(
    nn.Linear(4, 16),
    nn.ReLU(),
    nn.Linear(16, 3)
)

print("Model architecture:\n", model)
```

## 2 – Pass Input Data and Get Predictions

```python
# Create a sample input tensor (1 sample, 4 features)
input_data = torch.tensor([[0.5, -0.2, 1.0, 0.3]])

# Forward pass: Get raw output (logits)
output = model(input_data)

# Convert logits to class probabilities using softmax
probabilities = F.softmax(output, dim=1)

# Get predicted class (index of max probability)
predicted_class = torch.argmax(probabilities, dim=1)

print("\nModel output (logits):", output)
print("Class probabilities:", probabilities)
print("Predicted class:", predicted_class.item())
```

## 3 – Inspect Model Architecture, Weights, and Biases

```python
# Re‑print the model for reference
print("\nInspecting model structure:\n", model)

# Iterate through model layers to access weights and biases
for idx, layer in enumerate(model):
    if isinstance(layer, nn.Linear):
        print(f"\nLayer {idx}: {layer}")
        print("Weights:\n", layer.weight)
        print("Biases:\n", layer.bias)
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Activation function: Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Inputs to the neuron
x1 = 1.0
x2 = 2.0
x3 = 3.0
x4 = 4.0

# Weights and bias
w1 = 0.5
w2 = -1.0
w3 = 0.7
w4 = 1.2
b = 0.1

# Compute weighted sum (z)
z = (x1 * w1) + (x2 * w2)+(x3*w3)+(x4*w4) + b

# Output after activation
output = sigmoid(z)

# Display results
print(f"Input x1 = {x1}, x2 = {x2}, x3 = {x3}, x4 = {x4}")
print(f"Weights w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}")
print(f"Bias = {b}")
print(f"Weighted sum z = {z}")
print(f"Output after sigmoid activation = {output:.4f}")
```

```python
# Varying w1 and w2 to see how output changes
x1, x2 = 1.0, 2.0
bias = 0.1

w1_values = np.linspace(-2, 2, 50)
outputs = []

for w1 in w1_values:
    w2 = -1.0  # Keep w2 fixed
    z = (x1 * w1) + (x2 * w2) + bias
    out = sigmoid(z)
    outputs.append(out)

plt.plot(w1_values, outputs)
plt.title('Neuron Output vs Weight w1')
plt.xlabel('w1 value')
plt.ylabel('Neuron Output')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Input vector (1 sample, 4 features)
X = np.array([[1.0, 2.0, 3.0, 4.0]])  # Shape: (1, 4)

# Weight matrix for 16 neurons, each with 4 weights (4 inputs)
W = np.random.randn(4, 16)  # Shape: (4, 16)

# Bias for each of the 16 neurons
b = np.random.randn(1, 16)  # Shape: (1, 16)

# Compute weighted sum (Z) and apply sigmoid activation
Z = np.dot(X, W) + b        # Shape: (1, 16)
A = sigmoid(Z)              # Activated output (1, 16)

# Display
print("Input X (1x4):")
print(X)
print("\nWeight Matrix W (4x16):")
print(W)
print("\nBias b (1x16):")
print(b)
print("\nWeighted sum Z (1x16):")
print(Z)
print("\nOutput after sigmoid activation A (1x16):")
print(np.round(A, 4))
```
